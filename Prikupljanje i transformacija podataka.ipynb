###Transformacije tabele 1

Uklanjam kolone koje smatram suvišnim.

```
df1 = df1.drop("Timestamp", "Age")
#display(df1)
```

Ulepšavam imena kolona.

```
columnNames = ["Choose your gender", "What is your course?", "Your current year of Study", "What is your CGPA?", "Marital status", "Do you have Depression?", "Do you have Anxiety?", "Do you have Panic attack?", "Did you seek any specialist for a treatment?"]
newColumnNames = ["gender", "course", "year_of_study", "cgpa", "marital_status", "depression", "anxiety", "panic_attacks", "specialist_treatment"]

i = 0
for colName in newColumnNames:
    df1 = df1.withColumnRenamed(columnNames[i], colName)
    i += 1

df1.cache()
#display(df1)
```

Dovodim vrednosti kolone course u konzistentno stanje.

```
df1 = df1.withColumn("course", \
              when(df1.course == "BCS", "Computer Science") \
              .when(df1.course == "BIT", "Information Technology") \
              .when(df1.course == "IT", "Information Technology") \
              .when(df1.course == "Engineering", "Engineering") \
              .when(df1.course == "Engine", "Engineering") \
              .when(df1.course == "engin", "Engineering") \
              .when(df1.course == "ENM", "Engineering") \
              .when(df1.course == "KOE", "Engineering") \
              .when(df1.course == "Koe", "Engineering") \
              .when(df1.course == "koe", "Engineering") \
              .otherwise("Other"))
#display(df1)
```

Dovodim vrednosti kolone year_of_study u konzistentno stanje.

```
df1 = df1.withColumn("year_of_study", split(col("year_of_study")," ").getItem(1))
#display(df1)
```

###Transformacije tabele 2

Uklanjam kolone koje smatram suvišnim.

```
df2 = df2.drop("Timestamp", "state", "comments")
#display(df2)
```

Ulepšavam imena kolona.

```
columnNames = ["Age", "Gender", "Country", "self_employed", "family_history", "treatment", "work_interfere", "no_employees", "remote_work", "tech_company", "benefits", "care_options", "wellness_program", "seek_help", "anonymity", "leave", "mental_health_consequence", "phys_health_consequence", "coworkers", "supervisor", "mental_health_interview", "phys_health_interview", "mental_vs_physical", "obs_consequence"]
newColumnNames = ["age", "gender", "country", "self_employed", "family_history", "specialist_treatment", "interfering_with_work", "nr_of_employees", "working_remotely", "tech_company", "mental_health_benefits", "care_options", "wellness_program", "help_seeking_resources", "anonymity", "ease_of_medical_leave", "mental_health_consequence", "phys_health_consequence", "coworkers_trust", "supervisor_trust", "mental_health_interview", "phys_health_interview", "mental_vs_physical", "observed_consequences"]

i = 0
for colName in newColumnNames:
    df2 = df2.withColumnRenamed(columnNames[i], colName)
    i += 1

df2.cache()
#display(df2)
```

Dovodim vrednosti kolone age u konzistentno stanje.

```
df2 = df2.withColumn("age", \
              when(df2.age < 18, None) \
              .when(df2.age > 65, None) \
              .otherwise(df2.age))
#display(df2)
```

Dovodim vrednosti kolone gender u konzistentno stanje.

```
df2 = df2.withColumn("gender", \
              when(df2.gender == "msle", "Male") \
              .when(df2.gender == "Man", "Male") \
              .when(df2.gender == "Malr", "Male") \
              .when(df2.gender == "male", "Male") \
              .when(df2.gender == "Mal", "Male") \
              .when(df2.gender == "Make", "Male") \
              .when(df2.gender == "maile", "Male") \
              .when(df2.gender == "Mail", "Male") \
              .when(df2.gender == "M", "Male") \
              .when(df2.gender == "m", "Male") \
              .when(df2.gender == "Male", "Male") \
              .when(df2.gender == "Woman", "Female") \
              .when(df2.gender == "woman", "Female") \
              .when(df2.gender == "female", "Female") \
              .when(df2.gender == "Femake", "Female") \
              .when(df2.gender == "femail", "Female") \
              .when(df2.gender == "F", "Female") \
              .when(df2.gender == "f", "Female") \
              .when(df2.gender == "Female", "Female") \
              .otherwise("Other"))
#display(df2)
```

###Transformacije tabele 3

Uklanjam kolone koje smatram suvišnim.

```
df3 = df3.drop("index", "Code")
#display(df3)
```

Ulepšavam imena kolona.

```
columnNames = ["Entity", "Year", "Schizophrenia (%)", "Bipolar disorder (%)", "Eating disorders (%)", "Anxiety disorders (%)", "Drug use disorders (%)", "Depression (%)", "Alcohol use disorders (%)"]
newColumnNames = ["entity", "year", "schizophrenia_pct", "bipolar_disorder_pct", "eating_disorder_pct", "anxiety_pct", "drug_use_disorder_pct", "depression_pct", "alcohol_use_disorder_pct"]

i = 0
for colName in newColumnNames:
    df3 = df3.withColumnRenamed(columnNames[i], colName)
    i += 1

df3.cache()
#display(df3)
```

Zaokružujem sve numeričke vrednosti u tabeli na dve decimale.

```
df3 = df3.withColumn("schizophrenia_pct", round(col("schizophrenia_pct"), 2))
df3 = df3.withColumn("bipolar_disorder_pct", round(col("bipolar_disorder_pct"), 2))
df3 = df3.withColumn("eating_disorder_pct", round(col("eating_disorder_pct"), 2))
df3 = df3.withColumn("anxiety_pct", round(col("anxiety_pct"), 2))
df3 = df3.withColumn("drug_use_disorder_pct", round(col("drug_use_disorder_pct"), 2))
df3 = df3.withColumn("depression_pct", round(col("depression_pct"), 2))
df3 = df3.withColumn("alcohol_use_disorder_pct", round(col("alcohol_use_disorder_pct"), 2))
#display(df3)
```

###Transformacije tabele 4

Ulepšavam imena kolona.

```
columnNames = ["Country", "Sex", "2016", "2015", "2010", "2000"]
newColumnNames = ["country", "sex", "year_2016", "year_2015", "year_2010", "year_2000"]

i = 0
for colName in newColumnNames:
    df4 = df4.withColumnRenamed(columnNames[i], colName)
    i += 1

df4.cache()
#display(df4)
```

###Transformacije tabele 5

Iz osam kolona za starosne grupe stvaram tri. \
Starosne grupe "80 i više", "70-79", i "60-69" će biti spojene u kolonu "old_people". \
Starosne grupe "50-59", "40-49", i "30-39" će biti spojene u kolonu "middle_aged_people". \
Starosne grupe "20-29" i "10-19" će biti spojene u kolonu "young_people". \
Spajanje u nove kolone je uradjeno tako što se uzima prosek od vrednosti relevantnih kolona.

```
df5 = df5.withColumn("old_people", round((col(" 80_above") + col(" 70to79") + col(" 60to69 ")) / 3, 2))
df5 = df5.withColumn("middle_aged_people", round((col(" 50to59 ") + col(" 40to49") + col(" 30to39")) / 3, 2))
df5 = df5.withColumn("young_people", round((col(" 20to29") + col(" 10to19")) / 2, 2))
#display(df5)
```

Uklanjam stare kolone koju su mi sada suvišne, i ulepšavam imena ostalim kolonama.

```
df5 = df5.drop(" 80_above", " 70to79", " 60to69 ", " 50to59 ", " 40to49", " 30to39", " 20to29", " 10to19")
df5 = df5.withColumnRenamed("Country", "country")
df5 = df5.withColumnRenamed("Sex", "sex")
df5.cache()
#display(df5)
```
