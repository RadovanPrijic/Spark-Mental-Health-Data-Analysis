##1. Vizualizacija podataka

####1.1 Veza izmedju anksioznosti i visokog proseka kod studenata

```
%sql
select anxiety, 
count(*) as total, 
count(case when df1_view.cgpa = '3.50 - 4.00' then 1 end) as high_cgpa,
count(case when df1_view.cgpa != '3.50 - 4.00' then 1 end) as normal_cgpa
from df1_view
group by anxiety
order by anxiety desc
```

####1.2 Postotak studenata koji (ni)su zatra≈æili pomoƒá struƒçnjaka za mentalno zdravlje

```
%sql
select specialist_treatment, 
count(specialist_treatment) as number_of_students 
from df1_view  
group by specialist_treatment
order by specialist_treatment desc
```

####1.3 Prisutnost mentalnih tegoba kod ljudi koji rade u tehnolo≈°kom sektoru

```
%sql
select tech_company,
count(*) as total,
count(case when df2_view.interfering_with_work = 'Often' or df2_view.interfering_with_work = 'Sometimes' then 1 end) as mental_health_condition,
count(case when df2_view.interfering_with_work != 'Often' and df2_view.interfering_with_work != 'Sometimes' then 1 end) as no_mental_health_condition
from df2_view
group by tech_company
having tech_company = 'Yes'
```

####1.4 Veza izmedju porodiƒçne istorije mentalnih bolesti i trenutnog uticaja istih na osobin posao

```
%sql
select interfering_with_work, 
count(*) as total, 
count(case when df2_view.family_history = 'Yes' then 1 end) as has_family_history,
count(case when df2_view.family_history = 'No' then 1 end) as no_family_history
from df2_view
group by interfering_with_work
having interfering_with_work != 'NA'
order by interfering_with_work
```

####1.5 Prisutnost mentalnih tegoba kod ljudi koji rade od kuƒáe

```
%sql
select working_remotely,
count(*) as total,
count(case when df2_view.interfering_with_work = 'Often' or df2_view.interfering_with_work = 'Sometimes' then 1 end) as mental_health_condition,
count(case when df2_view.interfering_with_work != 'Often' and df2_view.interfering_with_work != 'Sometimes' then 1 end) as no_mental_health_condition
from df2_view
group by working_remotely
order by working_remotely desc
```

####1.6 Veza izmedju veliƒçine kompanije i ozbiljnosti shvaƒáanja mentalnih problema (u poredjenju sa telesnim)

```
%sql
select nr_of_employees,
count(*) as total,
count(case when df2_view.mental_vs_physical = 'Yes' then 1 end) as serious_mental_health_policy,
count(case when df2_view.mental_vs_physical = 'No' then 1 end) as lacking_mental_health_policy
from df2_view
group by nr_of_employees
order by nr_of_employees
```

####1.7 Poredjenje prisutnosti raznih mentalnih bolesti u ekonomski jakim zemljama sa svetom generalno

```
%sql
select entity, 
round(avg(anxiety_pct), 1) as avg_anxiety_pct, 
round(avg(depression_pct), 1) as avg_depression_pct,
round(avg(bipolar_disorder_pct), 1) as avg_bipolar_disorder_pct,
round(avg(schizophrenia_pct), 1) as avg_schizophrenia_pct,
round(avg(drug_use_disorder_pct), 1) as avg_drug_use_disorder_pct, 
round(avg(alcohol_use_disorder_pct), 1) as avg_alcohol_use_disorder_pct,
round(avg(eating_disorder_pct), 1) as avg_eating_disorder_pct
from df3_view
where anxiety_pct is not null 
and depression_pct is not null 
and bipolar_disorder_pct is not null 
and schizophrenia_pct is not null 
and drug_use_disorder_pct is not null 
and alcohol_use_disorder_pct is not null 
and eating_disorder_pct is not null
group by entity
having entity = 'World' or entity = 'High-income'
order by entity
```

####1.8 Kretanje postotka raznih mentalnih bolesti u Srbiji kroz vreme

```
%sql
select distinct year, 
anxiety_pct,
depression_pct, 
bipolar_disorder_pct,
schizophrenia_pct,
drug_use_disorder_pct, 
alcohol_use_disorder_pct,
eating_disorder_pct
from df3_view 
where entity = 'Serbia' 
and anxiety_pct is not null 
and depression_pct is not null 
and bipolar_disorder_pct is not null 
and schizophrenia_pct is not null 
and drug_use_disorder_pct is not null 
and alcohol_use_disorder_pct is not null 
and eating_disorder_pct is not null
order by year
```

####1.9 Proseƒçna stopa suicida u Srbiji kroz vreme (po polovima)

```
%sql
select sex, 
round(avg(year_2000), 2) as year_2000_avg_suicide_rate, 
round(avg(year_2010), 2) as year_2010_avg_suicide_rate, 
round(avg(year_2015), 2) as year_2015_avg_suicide_rate,
round(avg(year_2016), 2) as year_2016_avg_suicide_rate
from df4_view
where country = 'Serbia'
group by sex
order by sex desc
```

####1.10 Proseƒçna stopa suicida u Srbiji po starosnim grupama

```
%sql
select sex, 
round(avg(old_people), 2) as old_avg_suicide_rate, 
round(avg(middle_aged_people), 2) as middle_aged_avg_suicide_rate, 
round(avg(young_people), 2) as young_avg_suicide_rate
from df5_view
where country = 'Serbia'
group by sex
order by sex desc
```
